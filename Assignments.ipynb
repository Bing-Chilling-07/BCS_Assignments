## **Fibonacci Numbers**

```python
arr_fibb = [0] * 100                #creating an array of 100 elements each initialized to 0
arr_fibb[0] = 1
arr_fibb[1] = 1
for count in range(2, 100):
  arr_fibb[count] = arr_fibb[count - 1] + arr_fibb[count - 2]

print(arr_fibb)

```

## **Looping Through Numbers:**

```python
new_arr = np.arange(1, 51, 1)
for r in range (0, 50):
  if((new_arr[r]%3 == 0) & (new_arr[r]%5 != 0) ):              #when number is divisible by ONLY 3
    print("Fizz")
  elif((new_arr[r]%3 != 0) & (new_arr[r]%5 == 0) ):            #when number is divisible by ONLY 5
    print("Buzz")
  elif((new_arr[r]%3 == 0) & (new_arr[r]%5 == 0) ):            #when number is divisible by 3 AND 5
    print("FizzBuzz")
  else:                         #when number is not divisible by both 3 and 5
    print("LMAO")
```

## **Variable-Length Arguments:**

```python
def median(list):
  median = 0;     #using median as 0 if the length of the list provided is 0
  list.sort()
  n = len(list)
  if(n > 0):
   if(n%2 == 0):
    median = (list[(n-1)//2] + list[((n-1)//2) + 1])/2            #median formula for even-length list
   elif(n%2 != 0):
    median = list[n//2]                                           #median formula for odd-length list
  return median
```

```python
numbers = np.random.randint(1, 10, 6)
print(numbers)
numb_list = numbers.tolist()
print(median(numb_list))
```

```python
def average(lst):
    total = sum(lst)
    length = len(lst)
    if (length == 0):              #checking the "division by zero" case with if-else statements
        return "No Input Given"
    else:
        avg = total / length
        return avg

```

```python
user_input = input("Enter numbers: ")

numbers = list(map(int, user_input.split())) #taking user input as a string and converting it into a list using split() method

avg = average(numbers)
print(avg)

```

## **Error Handling:**

```python
#take the above example of calculating average
user_input_2 = input("Enter NO Number: ")
numbers_2 = list(map(int, user_input_2.split()))
avg_zero = average(numbers_2)
print(avg_zero)
```
